## 2. Test Score

```python

df2 = pd.read_csv("https://raw.githubusercontent.com/1stclasse3/DAV-5400/master/Test%20Score.csv")
df2
```

```python
# switch from wide to long format
melted2 = pd.melt(df2, id_vars=['Student'], value_vars=['Physics', 'Chemistry','English','Math'])
melted2
# change column name
rename2 = melted2.rename(index = str, columns = {"variable": "Class", "value": "Score"})
rename2
```

```python
# find max value per class
scoremax = rename2.groupby('Class').max()
scoremax.drop(columns=['Student'],inplace=True) # remove 'Student' column in order to show the score data
scoremax
```

```python
# find min value per class
scoremin =  rename2.groupby('Class').min()
scoremin.drop(columns=['Student'],inplace=True) # remove 'Student' column in order to show the score data
scoremin
```

Based on the above data analysis, I can get the highest score per class; Chemistry is 89, English is 91, Math is 95, and Physics is 78. Also, the lowest score per class is the following; Chemistry is 61, English is 56, Math is 64, and Physics is 22.

```python
plt.figure(figsize=(20,7))
sns.barplot(x='Student', y='Score', hue='Class', data=rename2); #Visualizing the result as a bar chart

```

This chart shows the overall scores per each student. Based on this chart, John has the highest scores in Math class and Alice has the highest scores in English class.
