```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/1stclasse3/DAV-5400/master/Data.csv")
df
```

```python
# change unnamed columns
rename = df.rename(index = str, columns = {"Unnamed: 0": "Airline", "Unnamed: 1": "Time"})
rename
```

```python
rename['Airline'].fillna(method='ffill', inplace=True) # fill out NA value 
rename
```

```python
rename1 = rename.dropna() # remove NaN data
rename1
```

```python
# switch from wide to long format
melted = pd.melt(rename1, id_vars=['Airline', 'Time'], value_vars=['Los Angeles', 'Phoenix', 'San Diego', 'San Francisco', 'Seattle'])
# change columns name
rename2 = melted.rename(index = str, columns = {"variable": "City", "value": "Minutes"})
rename2
```

## Q1. Which city had the best on time performance? 

```python
# find each city on time
eachcity = rename2[rename2['Time'] == 'on time'].groupby('City')['Minutes'].sum()
eachcity
```

```python
# include delayed time for each city
city = rename2.groupby('City')['Minutes'].sum()
city
```

```python
# calculate percentage
eachcity / city * 100
```

Based on the above result, I can say Phoenix airport has the best on time performance

## Q2. Which airline had the best overall on time performance?

```python
# find ontime per each airline
ontime = rename2[rename2['Time'] == 'on time'].groupby('Airline')['Minutes'].sum()
ontime
```

```python
# find delayed time per each airline
delayed = rename2[rename2['Time'] == 'delayed'].groupby('Airline')['Minutes'].sum()
delayed
```

```python
# calculate percentage and see which airline has the best performance
percentage = ontime / (ontime+delayed) * 100
percentage
```

Based on the result, AMWEST has the best overall on time performance.
